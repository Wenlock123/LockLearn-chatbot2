```python
from sentence_transformers import SentenceTransformer
import torch
import json
import numpy as np

# ğŸš€ à¹‚à¸«à¸¥à¸”à¹‚à¸¡à¹€à¸”à¸¥ SBERT (paraphrase-multilingual-mpnet-base-v2)
model_name = "sentence-transformers/paraphrase-multilingual-mpnet-base-v2"
model = SentenceTransformer(model_name)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
model.eval()  # à¸•à¸±à¹‰à¸‡à¹€à¸›à¹‡à¸™à¹‚à¸«à¸¡à¸”à¸›à¸£à¸°à¹€à¸¡à¸´à¸™à¸œà¸¥à¹€à¸à¸·à¹ˆà¸­à¸›à¸£à¸°à¸«à¸¢à¸±à¸”à¸«à¸™à¹ˆà¸§à¸¢à¸„à¸§à¸²à¸¡à¸ˆà¸³

# ğŸš€ à¸Ÿà¸±à¸‡à¸à¹Œà¸Šà¸±à¸™à¸ªà¸£à¹‰à¸²à¸‡ Embedding à¸‚à¸­à¸‡à¸„à¸³à¸–à¸²à¸¡
def embed_text_batch(texts):
    embeddings = model.encode(
        texts,
        convert_to_tensor=True,
        show_progress_bar=True,
        device=device,
        batch_size=32  # à¸ªà¸²à¸¡à¸²à¸£à¸–à¸›à¸£à¸±à¸š batch size à¹ƒà¸«à¹‰à¹€à¸«à¸¡à¸²à¸°à¸ªà¸¡à¸à¸±à¸š GPU à¹„à¸”à¹‰
    )
    return embeddings.cpu().numpy()

# ğŸš€ à¹‚à¸«à¸¥à¸” Evaluation Dataset
evaluation_file = '/content/drive/MyDrive/LockLearn/evaluation_datasetv2.json'
with open(evaluation_file, 'r', encoding='utf-8') as f:
    evaluation_data = json.load(f)

print(f"âœ… à¹‚à¸«à¸¥à¸” Evaluation Dataset à¸ˆà¸³à¸™à¸§à¸™à¸„à¸³à¸–à¸²à¸¡-à¸„à¸³à¸•à¸­à¸š: {len(evaluation_data)}")

# ğŸš€ à¸ªà¸£à¹‰à¸²à¸‡ Embeddings à¸ªà¸³à¸«à¸£à¸±à¸šà¸„à¸³à¸–à¸²à¸¡à¸—à¸±à¹‰à¸‡à¸«à¸¡à¸” (Batch)
batch_size = 10  # à¹à¸šà¹ˆà¸‡à¹€à¸›à¹‡à¸™ batch à¸¥à¸° 10 à¸„à¸³à¸–à¸²à¸¡
all_embeddings = []

for i in range(0, len(evaluation_data), batch_size):
    batch_questions = [item['question'] for item in evaluation_data[i:i + batch_size]]

    batch_embeddings = embed_text_batch(batch_questions)
    all_embeddings.extend(batch_embeddings.tolist())

    print(f"âœ… batch à¸—à¸µà¹ˆ {i // batch_size + 1} embed à¹à¸¥à¹‰à¸§")

    # ğŸš€ à¹€à¸„à¸¥à¸µà¸¢à¸£à¹Œ Cache à¸‚à¸­à¸‡ GPU à¹€à¸à¸·à¹ˆà¸­à¸›à¸£à¸°à¸«à¸¢à¸±à¸”à¸«à¸™à¹ˆà¸§à¸¢à¸„à¸§à¸²à¸¡à¸ˆà¸³
    if device.type == "cuda":
        torch.cuda.empty_cache()

# ğŸš€ à¹€à¸à¸´à¹ˆà¸¡ Embeddings à¸à¸¥à¸±à¸šà¹€à¸‚à¹‰à¸² Evaluation Dataset
for i, item in enumerate(evaluation_data):
    item['embedding'] = all_embeddings[i]

print("âœ… à¸ªà¸£à¹‰à¸²à¸‡ Embeddings à¸ªà¸³à¸«à¸£à¸±à¸šà¸„à¸³à¸–à¸²à¸¡à¹ƒà¸™ Evaluation Dataset à¸ªà¸³à¹€à¸£à¹‡à¸ˆ")

# ğŸš€ à¸à¸³à¸«à¸™à¸”à¹€à¸ªà¹‰à¸™à¸—à¸²à¸‡à¸—à¸µà¹ˆà¸•à¹‰à¸­à¸‡à¸à¸²à¸£à¸šà¸±à¸™à¸—à¸¶à¸à¹„à¸Ÿà¸¥à¹Œ (Google Drive)
output_file = '/content/drive/MyDrive/LockLearn/evaluation_dataset_with_embeddings_v2_SBERT.json'

# ğŸš€ à¸šà¸±à¸™à¸—à¸¶à¸ Evaluation Dataset à¸à¸£à¹‰à¸­à¸¡ Embeddings à¸¥à¸‡à¹„à¸Ÿà¸¥à¹Œ JSON
with open(output_file, 'w', encoding='utf-8') as f:
    json.dump(evaluation_data, f, ensure_ascii=False, indent=4)

print(f"âœ… à¸šà¸±à¸™à¸—à¸¶à¸ Evaluation Dataset à¸à¸£à¹‰à¸­à¸¡ Embeddings à¸—à¸µà¹ˆ: {output_file}")
```

à¸«à¸²à¸„à¹ˆà¸² Recall

```python
import json

# ğŸš€ à¹‚à¸«à¸¥à¸”à¹„à¸Ÿà¸¥à¹Œ evaluation dataset (à¸„à¸³à¸–à¸²à¸¡-à¸„à¸³à¸•à¸­à¸šà¸ˆà¸£à¸´à¸‡)
evaluation_file = '/content/drive/MyDrive/LockLearn/evaluation_dataset_with_embeddings_v2_SBERT.json'
with open(evaluation_file, 'r', encoding='utf-8') as f:
    evaluation_data = json.load(f)

# ğŸš€ à¹‚à¸«à¸¥à¸”à¹„à¸Ÿà¸¥à¹Œà¸œà¸¥à¸¥à¸±à¸à¸˜à¹Œ RAG (à¸„à¸³à¸–à¸²à¸¡ + à¸„à¸³à¹à¸™à¸°à¸™à¸³à¸—à¸µà¹ˆà¸”à¸¶à¸‡à¸¡à¸²)
rag_results_file = '/content/drive/MyDrive/LockLearn/evaluation_rag_results_v2_SBERT_K=3.json'
with open(rag_results_file, 'r', encoding='utf-8') as f:
    rag_results = json.load(f)

K = 3  # à¸ˆà¸³à¸™à¸§à¸™à¸„à¸³à¹à¸™à¸°à¸™à¸³à¸—à¸µà¹ˆà¸”à¸¶à¸‡à¸¡à¸² (top_k)

# ğŸš€ à¸Ÿà¸±à¸‡à¸à¹Œà¸Šà¸±à¸™à¸„à¸³à¸™à¸§à¸“ Precision@K
def precision_at_k(relevant, retrieved, k):
    retrieved_k = retrieved[:k]
    hits = sum(1 for r in retrieved_k if r in relevant)
    return hits / k if k > 0 else 0

# ğŸš€ à¸Ÿà¸±à¸‡à¸à¹Œà¸Šà¸±à¸™à¸„à¸³à¸™à¸§à¸“ Recall@K
def recall_at_k(relevant, retrieved, k):
    retrieved_k = retrieved[:k]
    hits = sum(1 for r in retrieved_k if r in relevant)
    return hits / len(relevant) if relevant else 0

# ğŸš€ à¸Ÿà¸±à¸‡à¸à¹Œà¸Šà¸±à¸™à¸„à¸³à¸™à¸§à¸“ Mean Reciprocal Rank (MRR)
def mean_reciprocal_rank(relevant, retrieved):
    for idx, rec in enumerate(retrieved, 1):
        if rec in relevant:
            return 1 / idx
    return 0

# ğŸš€ à¹€à¸à¹‡à¸šà¸œà¸¥à¸¥à¸±à¸à¸˜à¹Œà¸„à¸³à¸™à¸§à¸“à¸•à¹ˆà¸²à¸‡à¹†
precisions = []
recalls = []
mrrs = []

print("ğŸ”„ à¸à¸³à¸¥à¸±à¸‡à¸„à¸³à¸™à¸§à¸“ Precision, Recall, à¹à¸¥à¸° MRR...")

for idx, item in enumerate(rag_results, 1):
    # à¸•à¸£à¸§à¸ˆà¸ªà¸­à¸šà¸§à¹ˆà¸²à¸¡à¸µà¸„à¸³à¸•à¸­à¸šà¹à¸¥à¸°à¸„à¸³à¹à¸™à¸°à¸™à¸³à¸«à¸£à¸·à¸­à¹„à¸¡à¹ˆ
    if 'answer' not in item or 'recommendations' not in item:
        print(f"âš ï¸ à¸‚à¹‰à¸­à¸—à¸µà¹ˆ {idx} à¹„à¸¡à¹ˆà¸¡à¸µà¸„à¸³à¸•à¸­à¸šà¸«à¸£à¸·à¸­à¸„à¸³à¹à¸™à¸°à¸™à¸³ à¸‚à¹‰à¸²à¸¡à¹„à¸›...")
        continue

    # à¸„à¸³à¸•à¸­à¸šà¸ˆà¸£à¸´à¸‡ (à¹ƒà¸™à¸£à¸¹à¸›à¹à¸šà¸š list à¹€à¸à¸·à¹ˆà¸­à¸£à¸­à¸‡à¸£à¸±à¸šà¸«à¸¥à¸²à¸¢à¸„à¸³à¸•à¸­à¸šà¹ƒà¸™à¸­à¸™à¸²à¸„à¸•)
    relevant_answers = [item['answer']] if isinstance(item['answer'], str) else item['answer']

    # à¸„à¸³à¹à¸™à¸°à¸™à¸³à¸—à¸µà¹ˆà¸”à¸¶à¸‡à¸¡à¸² (list of strings)
    retrieved = item['recommendations']

    # à¸„à¸³à¸™à¸§à¸“ Metrics
    p = precision_at_k(relevant_answers, retrieved, K)
    r = recall_at_k(relevant_answers, retrieved, K)
    mrr = mean_reciprocal_rank(relevant_answers, retrieved)

    precisions.append(p)
    recalls.append(r)
    mrrs.append(mrr)

    if idx % 100 == 0 or idx == len(rag_results):
        print(f"âœ… à¸›à¸£à¸°à¸¡à¸§à¸¥à¸œà¸¥à¹à¸¥à¹‰à¸§ {idx}/{len(rag_results)} à¸„à¸³à¸–à¸²à¸¡")

# ğŸš€ à¸„à¸³à¸™à¸§à¸“à¸„à¹ˆà¸²à¹€à¸‰à¸¥à¸µà¹ˆà¸¢à¸‚à¸­à¸‡ Precision, Recall, MRR à¸—à¸±à¹‰à¸‡à¸«à¸¡à¸”
avg_precision = sum(precisions) / len(precisions) if precisions else 0
avg_recall = sum(recalls) / len(recalls) if recalls else 0
avg_mrr = sum(mrrs) / len(mrrs) if mrrs else 0

print("\nğŸ” à¸œà¸¥à¸¥à¸±à¸à¸˜à¹Œà¸à¸²à¸£à¸›à¸£à¸°à¹€à¸¡à¸´à¸™à¸œà¸¥à¸”à¹‰à¸§à¸¢ RAG")
print(f"Precision@{K}: {avg_precision:.3f}")
print(f"Recall@{K}: {avg_recall:.3f}")
print(f"Mean Reciprocal Rank (MRR): {avg_mrr:.3f}")
```

```python
import torch
from transformers import AutoTokenizer, AutoModel
import json

# ğŸ”§ à¸à¸³à¸«à¸™à¸”à¹€à¸ªà¹‰à¸™à¸—à¸²à¸‡à¹„à¸Ÿà¸¥à¹Œ
input_file = '/content/drive/MyDrive/LockLearn/evaluation_dataset_TH.json'
output_file = '/content/drive/MyDrive/LockLearn/evaluation_dataset_TH_embedded_SBERT.json'

# ğŸš€ à¹‚à¸«à¸¥à¸”à¹‚à¸¡à¹€à¸”à¸¥ Hugging Face (paraphrase-multilingual-mpnet-base-v2)
model_name = 'sentence-transformers/paraphrase-multilingual-mpnet-base-v2'
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
model.eval()  # à¸•à¸±à¹‰à¸‡à¹‚à¸«à¸¡à¸”à¸›à¸£à¸°à¹€à¸¡à¸´à¸™à¸œà¸¥à¹€à¸à¸·à¹ˆà¸­à¸›à¸£à¸°à¸«à¸¢à¸±à¸”à¸«à¸™à¹ˆà¸§à¸¢à¸„à¸§à¸²à¸¡à¸ˆà¸³

# ğŸš€ à¹‚à¸«à¸¥à¸”à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸ˆà¸²à¸à¹„à¸Ÿà¸¥à¹Œ
with open(input_file, 'r', encoding='utf-8') as f:
    data = json.load(f)

# ğŸ”¥ à¸Ÿà¸±à¸‡à¸à¹Œà¸Šà¸±à¸™à¸ªà¸£à¹‰à¸²à¸‡ Embedding
def create_embedding(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
    with torch.no_grad():
        embeddings = model(**inputs).last_hidden_state.mean(dim=1)
    return embeddings[0].cpu().numpy().tolist()

# âœ… à¹€à¸£à¸´à¹ˆà¸¡à¸ªà¸£à¹‰à¸²à¸‡ Embedding
print(f"âœ… à¸£à¸§à¸¡à¸„à¸³à¸–à¸²à¸¡à¸—à¸±à¹‰à¸‡à¸«à¸¡à¸”: {len(data)}")

for index, item in enumerate(data, start=1):
    text = item.get("translated_question", "")
    if text:
        item['question_embedding'] = create_embedding(text)
        print(f"âœ… à¸„à¸³à¸–à¸²à¸¡à¸—à¸µà¹ˆ {index} à¹„à¸”à¹‰à¸£à¸±à¸šà¸à¸²à¸£à¸ªà¸£à¹‰à¸²à¸‡ Embedding à¹à¸¥à¹‰à¸§")

# ğŸ’¾ à¸šà¸±à¸™à¸—à¸¶à¸à¸œà¸¥à¸¥à¸±à¸à¸˜à¹Œà¸—à¸±à¹‰à¸‡à¸«à¸¡à¸”à¸«à¸¥à¸±à¸‡à¸›à¸£à¸°à¸¡à¸§à¸¥à¸œà¸¥
with open(output_file, 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=4)

print("âœ… à¹€à¸ªà¸£à¹‡à¸ˆà¸ªà¸´à¹‰à¸™à¸à¸²à¸£à¸ªà¸£à¹‰à¸²à¸‡ Embedding à¸—à¸±à¹‰à¸‡à¸«à¸¡à¸”")
```

```python
import json

# ğŸš€ à¹‚à¸«à¸¥à¸”à¹„à¸Ÿà¸¥à¹Œ evaluation dataset (à¸„à¸³à¸–à¸²à¸¡-à¸„à¸³à¸•à¸­à¸šà¸ˆà¸£à¸´à¸‡)
evaluation_file = '/content/drive/MyDrive/LockLearn/evaluation_dataset_with_embeddings_v2_SBERT.json'
with open(evaluation_file, 'r', encoding='utf-8') as f:
    evaluation_data = json.load(f)

# ğŸš€ à¹‚à¸«à¸¥à¸”à¹„à¸Ÿà¸¥à¹Œà¸œà¸¥à¸¥à¸±à¸à¸˜à¹Œ RAG (à¸„à¸³à¸–à¸²à¸¡ + à¸„à¸³à¹à¸™à¸°à¸™à¸³à¸—à¸µà¹ˆà¸”à¸¶à¸‡à¸¡à¸²)
rag_results_file = '/content/drive/MyDrive/LockLearn/evaluation_rag_results_v2_TH_SBERT_K=3.json'  # à¹à¸à¹‰à¸Šà¸·à¹ˆà¸­à¹„à¸Ÿà¸¥à¹Œà¸•à¸²à¸¡à¸—à¸µà¹ˆà¸šà¸±à¸™à¸—à¸¶à¸à¹„à¸§à¹‰à¸à¹ˆà¸­à¸™à¸«à¸™à¹‰à¸²
with open(rag_results_file, 'r', encoding='utf-8') as f:
    rag_results = json.load(f)

K = 3  # à¸ˆà¸³à¸™à¸§à¸™à¸„à¸³à¹à¸™à¸°à¸™à¸³à¸—à¸µà¹ˆà¸”à¸¶à¸‡à¸¡à¸² (top_k)

# ğŸš€ à¸Ÿà¸±à¸‡à¸à¹Œà¸Šà¸±à¸™à¸„à¸³à¸™à¸§à¸“ Precision@K
def precision_at_k(relevant, retrieved, k):
    retrieved_k = retrieved[:k]
    hits = sum(1 for r in retrieved_k if r in relevant)
    return hits / k if k > 0 else 0

# ğŸš€ à¸Ÿà¸±à¸‡à¸à¹Œà¸Šà¸±à¸™à¸„à¸³à¸™à¸§à¸“ Recall@K
def recall_at_k(relevant, retrieved, k):
    retrieved_k = retrieved[:k]
    hits = sum(1 for r in retrieved_k if r in relevant)
    return hits / len(relevant) if relevant else 0

# ğŸš€ à¸Ÿà¸±à¸‡à¸à¹Œà¸Šà¸±à¸™à¸„à¸³à¸™à¸§à¸“ Mean Reciprocal Rank (MRR)
def mean_reciprocal_rank(relevant, retrieved):
    for idx, rec in enumerate(retrieved, 1):
        if rec in relevant:
            return 1 / idx
    return 0

# ğŸš€ à¹€à¸à¹‡à¸šà¸œà¸¥à¸¥à¸±à¸à¸˜à¹Œà¸„à¸³à¸™à¸§à¸“à¸•à¹ˆà¸²à¸‡à¹†
precisions = []
recalls = []
mrrs = []

print("ğŸ”„ à¸à¸³à¸¥à¸±à¸‡à¸„à¸³à¸™à¸§à¸“ Precision, Recall, à¹à¸¥à¸° MRR...")

for idx, item in enumerate(rag_results, 1):
    # à¸•à¸£à¸§à¸ˆà¸ªà¸­à¸šà¸§à¹ˆà¸²à¸¡à¸µà¸„à¸³à¸•à¸­à¸šà¹à¸¥à¸°à¸„à¸³à¹à¸™à¸°à¸™à¸³à¸«à¸£à¸·à¸­à¹„à¸¡à¹ˆ
    if 'answer' not in item or 'recommendations' not in item:
        print(f"âš ï¸ à¸‚à¹‰à¸­à¸—à¸µà¹ˆ {idx} à¹„à¸¡à¹ˆà¸¡à¸µà¸„à¸³à¸•à¸­à¸šà¸«à¸£à¸·à¸­à¸„à¸³à¹à¸™à¸°à¸™à¸³ à¸‚à¹‰à¸²à¸¡à¹„à¸›...")
        continue

    # à¸„à¸³à¸•à¸­à¸šà¸ˆà¸£à¸´à¸‡ (à¸£à¸­à¸‡à¸£à¸±à¸šà¸—à¸±à¹‰à¸‡ str à¸«à¸£à¸·à¸­ list)
    relevant_answers = [item['answer']] if isinstance(item['answer'], str) else item['answer']

    # à¸„à¸³à¹à¸™à¸°à¸™à¸³à¸—à¸µà¹ˆà¸”à¸¶à¸‡à¸¡à¸² (list of strings)
    retrieved = item['recommendations']

    # à¸„à¸³à¸™à¸§à¸“ Metrics
    p = precision_at_k(relevant_answers, retrieved, K)
    r = recall_at_k(relevant_answers, retrieved, K)
    mrr = mean_reciprocal_rank(relevant_answers, retrieved)

    precisions.append(p)
    recalls.append(r)
    mrrs.append(mrr)

    if idx % 100 == 0 or idx == len(rag_results):
        print(f"âœ… à¸›à¸£à¸°à¸¡à¸§à¸¥à¸œà¸¥à¹à¸¥à¹‰à¸§ {idx}/{len(rag_results)} à¸„à¸³à¸–à¸²à¸¡")

# ğŸš€ à¸„à¸³à¸™à¸§à¸“à¸„à¹ˆà¸²à¹€à¸‰à¸¥à¸µà¹ˆà¸¢à¸‚à¸­à¸‡ Precision, Recall, MRR à¸—à¸±à¹‰à¸‡à¸«à¸¡à¸”
avg_precision = sum(precisions) / len(precisions) if precisions else 0
avg_recall = sum(recalls) / len(recalls) if recalls else 0
avg_mrr = sum(mrrs) / len(mrrs) if mrrs else 0

print("\nğŸ” à¸œà¸¥à¸¥à¸±à¸à¸˜à¹Œà¸à¸²à¸£à¸›à¸£à¸°à¹€à¸¡à¸´à¸™à¸œà¸¥à¸”à¹‰à¸§à¸¢ RAG")
print(f"Precision@{K}: {avg_precision:.3f}")
print(f"Recall@{K}: {avg_recall:.3f}")
print(f"Mean Reciprocal Rank (MRR): {avg_mrr:.3f}")
```

```python
import json

# ğŸš€ à¹‚à¸«à¸¥à¸”à¹„à¸Ÿà¸¥à¹Œ evaluation dataset (à¸„à¸³à¸–à¸²à¸¡-à¸„à¸³à¸•à¸­à¸šà¸ˆà¸£à¸´à¸‡)
evaluation_file = '/content/drive/MyDrive/LockLearn/evaluation_dataset_with_embeddings_v2_SBERT.json'
with open(evaluation_file, 'r', encoding='utf-8') as f:
    evaluation_data = json.load(f)

# ğŸš€ à¹‚à¸«à¸¥à¸”à¹„à¸Ÿà¸¥à¹Œà¸œà¸¥à¸¥à¸±à¸à¸˜à¹Œ RAG (à¸„à¸³à¸–à¸²à¸¡ + à¸„à¸³à¹à¸™à¸°à¸™à¸³à¸—à¸µà¹ˆà¸”à¸¶à¸‡à¸¡à¸²)
rag_results_file = '/content/drive/MyDrive/LockLearn/evaluation_rag_results_v2_TH_SBERT_K=5.json'  # à¹à¸à¹‰à¸Šà¸·à¹ˆà¸­à¹„à¸Ÿà¸¥à¹Œà¸•à¸²à¸¡à¸—à¸µà¹ˆà¸šà¸±à¸™à¸—à¸¶à¸à¹„à¸§à¹‰à¸à¹ˆà¸­à¸™à¸«à¸™à¹‰à¸²
with open(rag_results_file, 'r', encoding='utf-8') as f:
    rag_results = json.load(f)

K = 5  # à¸ˆà¸³à¸™à¸§à¸™à¸„à¸³à¹à¸™à¸°à¸™à¸³à¸—à¸µà¹ˆà¸”à¸¶à¸‡à¸¡à¸² (top_k)

# ğŸš€ à¸Ÿà¸±à¸‡à¸à¹Œà¸Šà¸±à¸™à¸„à¸³à¸™à¸§à¸“ Precision@K
def precision_at_k(relevant, retrieved, k):
    retrieved_k = retrieved[:k]
    hits = sum(1 for r in retrieved_k if r in relevant)
    return hits / k if k > 0 else 0

# ğŸš€ à¸Ÿà¸±à¸‡à¸à¹Œà¸Šà¸±à¸™à¸„à¸³à¸™à¸§à¸“ Recall@K
def recall_at_k(relevant, retrieved, k):
    retrieved_k = retrieved[:k]
    hits = sum(1 for r in retrieved_k if r in relevant)
    return hits / len(relevant) if relevant else 0

# ğŸš€ à¸Ÿà¸±à¸‡à¸à¹Œà¸Šà¸±à¸™à¸„à¸³à¸™à¸§à¸“ Mean Reciprocal Rank (MRR)
def mean_reciprocal_rank(relevant, retrieved):
    for idx, rec in enumerate(retrieved, 1):
        if rec in relevant:
            return 1 / idx
    return 0

# ğŸš€ à¹€à¸à¹‡à¸šà¸œà¸¥à¸¥à¸±à¸à¸˜à¹Œà¸„à¸³à¸™à¸§à¸“à¸•à¹ˆà¸²à¸‡à¹†
precisions = []
recalls = []
mrrs = []

print("ğŸ”„ à¸à¸³à¸¥à¸±à¸‡à¸„à¸³à¸™à¸§à¸“ Precision, Recall, à¹à¸¥à¸° MRR...")

for idx, item in enumerate(rag_results, 1):
    # à¸•à¸£à¸§à¸ˆà¸ªà¸­à¸šà¸§à¹ˆà¸²à¸¡à¸µà¸„à¸³à¸•à¸­à¸šà¹à¸¥à¸°à¸„à¸³à¹à¸™à¸°à¸™à¸³à¸«à¸£à¸·à¸­à¹„à¸¡à¹ˆ
    if 'answer' not in item or 'recommendations' not in item:
        print(f"âš ï¸ à¸‚à¹‰à¸­à¸—à¸µà¹ˆ {idx} à¹„à¸¡à¹ˆà¸¡à¸µà¸„à¸³à¸•à¸­à¸šà¸«à¸£à¸·à¸­à¸„à¸³à¹à¸™à¸°à¸™à¸³ à¸‚à¹‰à¸²à¸¡à¹„à¸›...")
        continue

    # à¸„à¸³à¸•à¸­à¸šà¸ˆà¸£à¸´à¸‡ (à¸£à¸­à¸‡à¸£à¸±à¸šà¸—à¸±à¹‰à¸‡ str à¸«à¸£à¸·à¸­ list)
    relevant_answers = [item['answer']] if isinstance(item['answer'], str) else item['answer']

    # à¸„à¸³à¹à¸™à¸°à¸™à¸³à¸—à¸µà¹ˆà¸”à¸¶à¸‡à¸¡à¸² (list of strings)
    retrieved = item['recommendations']

    # à¸„à¸³à¸™à¸§à¸“ Metrics
    p = precision_at_k(relevant_answers, retrieved, K)
    r = recall_at_k(relevant_answers, retrieved, K)
    mrr = mean_reciprocal_rank(relevant_answers, retrieved)

    precisions.append(p)
    recalls.append(r)
    mrrs.append(mrr)

    if idx % 100 == 0 or idx == len(rag_results):
        print(f"âœ… à¸›à¸£à¸°à¸¡à¸§à¸¥à¸œà¸¥à¹à¸¥à¹‰à¸§ {idx}/{len(rag_results)} à¸„à¸³à¸–à¸²à¸¡")

# ğŸš€ à¸„à¸³à¸™à¸§à¸“à¸„à¹ˆà¸²à¹€à¸‰à¸¥à¸µà¹ˆà¸¢à¸‚à¸­à¸‡ Precision, Recall, MRR à¸—à¸±à¹‰à¸‡à¸«à¸¡à¸”
avg_precision = sum(precisions) / len(precisions) if precisions else 0
avg_recall = sum(recalls) / len(recalls) if recalls else 0
avg_mrr = sum(mrrs) / len(mrrs) if mrrs else 0

print("\nğŸ” à¸œà¸¥à¸¥à¸±à¸à¸˜à¹Œà¸à¸²à¸£à¸›à¸£à¸°à¹€à¸¡à¸´à¸™à¸œà¸¥à¸”à¹‰à¸§à¸¢ RAG")
print(f"Precision@{K}: {avg_precision:.3f}")
print(f"Recall@{K}: {avg_recall:.3f}")
print(f"Mean Reciprocal Rank (MRR): {avg_mrr:.3f}")
```
